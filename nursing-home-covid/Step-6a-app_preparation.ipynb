```python
def nursing_info(var1):
"""This it the concept code for the app at it's base functionality. It checks to see if the zipcode is in the nursing data. 
    If it is it returns those homes otherwise it returns the homes in the same state as that zip"""

    if len(df[df['Provider Zip Code']==zipcode]) > 0:
        return df[df['Provider Zip Code']==zipcode]
    
    else:  
        state = lookup[lookup['zip']==zipcode]['Provider State'].values
        return df[df['Provider State']==state[0]]
```
